##### <mark>3 - Inizializza l'istanza chat_model di LangChain che fornisce un'interfaccia per richiamare un provider LLM tramite chat API. Il motivo per cui selezioniamo il modello</mark>
##### <mark>3chat è che il modello gpt-35-turbo è ottimizzato per la chat, quindi utilizziamo la classe AzureChatOpenAI qui per inizializzare l'istanza.</mark>

```python
import os
from langchain.chat_models import AzureChatOpenAI

llm = AzureChatOpenAI(model=os.getenv("OPENAI_CHAT_MODEL"),
                      deployment_name=os.getenv("OPENAI_CHAT_MODEL"),
                      temperature=0)
```

##### <mark>4 - create prompt  template </mark>
<mark>4 Prompt is the input that we send to the LLM to generate an output. Prompt can also be designed to contain instructions, context, examples (one shot or few shot) which can be crucial for generating accurate output, as well as setting the tone and formatting your output data.</mark>

<mark>4 Using Prompt Template is a good way to structure these properties including the end user’s input to be provided to the LLM. We use LangChain’s ChatPromptTemplate module here, which is based on ChatML (Chat Markup Language).</mark>

<mark>4 This is a base prompt template to start with. Over time, we will update this template as needed —</mark>

```python
from langchain.prompts.chat import ChatPromptTemplate

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", 
         """
          You are a helpful AI assistant expert in querying SQL Database to find answers to user's question about Collision.
         """
         ),
        ("user", "{question}\n ai: "),
    ]
)
```

##### <mark>5 - Ora inizializza il create_sql_agent che è progettato per interagire con il database SQL come di seguito. L'agente è dotato di toolkit per connettersi al tuo database SQL e leggere sia i metadati che il contenuto delle tabelle.</mark>

```python
from langchain.agents import AgentType, create_sql_agent
from langchain.sql_database import SQLDatabase
from langchain.agents.agent_toolkits.sql.toolkit import SQLDatabaseToolkit

db = SQLDatabase(db_engine, view_support=True)
# test the connection
print(db.dialect)
print(db.get_usable_table_names())

sql_toolkit = SQLDatabaseToolkit(db=db, llm=llm)
sql_toolkit.get_tools()

sqldb_agent = create_sql_agent(
    llm=llm,
    toolkit=sql_toolkit,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

Controllo se il database è istanziato correttamente

```python
sqldb_agent.run(final_prompt.format(
        question="what is total of persone morte"
  ))
  #  question="what is the number of persone morte in the year 2023?"
```
